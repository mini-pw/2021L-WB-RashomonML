```python
sns.ecdfplot(mean_test_score, stat="count")
plt.title('Distribution of mean test score results')
plt.show()
```

### Rashomon ratio w zależności od punktu odcięcia

```python
counts, bins = np.histogram(mean_test_score, bins=50, range=(0.5, 1))
```

```python
reverse_cum_sum = np.cumsum(counts[::-1])[::-1] 
```

```python
ratio = reverse_cum_sum / len(mean_test_score)
ratio = np.append(ratio, 0)
```

```python
sns.lineplot(x = bins, y = ratio)
plt.title('Rashomon ratio')
plt.xlabel('Mean test score')
plt.ylabel('Rashomon ratio')
plt.show()
```

Najwięcej wyników jest w przedziale $0.8 - 0.85$, czyli zostało wygenerowanych dużo dobrych modeli - najlepszy model osiągnął wynik około $0.87$.

Zauważmy, że na wykresie *Rashomon ratio* linia ma największy spadek również w tym przedziale.

Ponadto, krzywa jest dość ładnie wygładzona w okolicach punktu $0.75$, kiedy "przekrzywia się" ku dołowi.

Krzywa następnie dość mocno spada w dół i w okolicach punktu $0.87$ pojawia się ostrze. Można przypuszczać, że trudno (a może niemożliwe) jest polepszenie wyników w takiej reprezentacji cech.
